```python
!java -Xmx8g -jar snpEff/snpEff.jar -v GRCh38.99 chd_child.vcf > chd_child.ann.vcf
```

```python
!egrep "#|synonymous" chd_child.ann.vcf | wc -l
```

```python
!egrep "#|synonymous|missense|stop_gained" chd_child.ann.vcf > chd_child.ann.coding.vcf 
```

```python
! java -jar snpEff/SnpSift.jar tstv  chd_child.ann.coding.vcf 
```

```python
!gatk SelectVariants \
--variant chd_child.ann.coding.vcf \
--select-type SNP \
--output chd_child.ann.coding.SNP.vcf
```

```python
!gatk SelectVariants \
--variant chd_child.ann.coding.vcf \
--select-type INDEL \
--output chd_child.ann.coding.INDELs.vcf
```

```python
!gatk VariantFiltration \
--variant chd_child.ann.coding.SNP.vcf \
--filter-expression "QD < 2.0" --filter-name "QD2" \
--filter-expression "QUAL < 30.0" --filter-name "QUAL30" \
--filter-expression "SOR > 3.0" --filter-name "SOR3" \
--filter-expression "FS > 60.0" --filter-name "FS60" \
--filter-expression "MQ < 40.0" --filter-name "MQ40" \
--output chd_child.ann.coding.SNP.filtered.vcf
```

```python
!gatk VariantFiltration \
--variant chd_child.ann.coding.INDELs.vcf \
--filter-expression "QD < 2.0" --filter-name "QD2" \
--filter-expression "QUAL < 30.0" --filter-name "QUAL30" \
--filter-expression "FS > 200.0" --filter-name "FS200" \
--output chd_child.ann.coding.INDELs.filtered.vcf
```

```python
!gatk MergeVcfs \
--INPUT chd_child.ann.coding.SNP.filtered.vcf \
--INPUT chd_child.ann.coding.INDELs.filtered.vcf \
--OUTPUT chd_child.ann.coding.filtered.vcf
```
